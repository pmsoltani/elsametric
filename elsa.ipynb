```python
from elsapy.elsclient import ElsClient
from elsapy.elsprofile import ElsAuthor, ElsAffil
from elsapy.elsdoc import FullDoc, AbsDoc
from elsapy.elssearch import ElsSearch
import json
```

```python
## Load configuration
con_file = open("config.json")
config = json.load(con_file)
con_file.close()

## Initialize client
client = ElsClient(config['apikey'])
client.inst_token = config['insttoken']
```

```python
## Author example
# Initialize author with uri
my_auth = ElsAuthor(
        uri = 'https://api.elsevier.com/content/author/author_id/15036893200')
# Read author data, then write to disk
if my_auth.read(client):
    print ("my_auth.full_name: ", my_auth.full_name)
else:
    print ("Read author failed.")
```

```python
doc_srch = ElsSearch('AFFIL(university AND of AND tehran) AND AFFILCITY(tehran) AND AFFILCOUNTRY(iran) AND (PUBYEAR > 2018)','scopus')
doc_srch.execute(els_client=client, get_all=True)
# doc_srch.execute(client, get_all = True)
print ("doc_srch has", len(doc_srch.results), "results.")
```

```python
## Scopus (Abtract) document example
# Initialize document with Scopus ID.
scp_doc = AbsDoc(scp_id = '85063494084')
if scp_doc.read(client):
    print ("scp_doc.title: ", scp_doc.title)  
else:
    print ("Read document failed.")
```
